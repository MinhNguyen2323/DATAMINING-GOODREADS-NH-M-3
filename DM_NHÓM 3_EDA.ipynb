```python
import pandas as pd
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
```

```python
df_review = pd.read_csv('/content/goodreads_reviews.csv')
df_book = pd.read_csv('/content/goodreads_works.csv')
```

```python
author_counts = df_book['author'].value_counts()

plt.figure(figsize=(12, 6))
sbn.barplot(x=author_counts.head(10).index, y=author_counts.head(10).values, palette='viridis')
plt.title('Top 10 tc gi c nhiu sch c revew nht')
plt.xlabel('Author')
plt.ylabel('Number of Books')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
title_info = df_book.groupby('original_title').agg(
    work_ids=('work_id', lambda x: list(x.unique())),
    authors=('author', lambda x: list(x.unique())),
    title_count=('original_title', 'count')
).sort_values(by='title_count', ascending=False)

title_info
```

```python
df_book['genres_list'] = (
    df_book['genres'].str.split(',')
    .apply(lambda x: [g.strip() for g in x])
)

df_genre = df_book.explode('genres_list')
```

```python
genre_count = df_genre['genres_list'].value_counts()

plt.figure(figsize=(12, 6))
sbn.barplot(x=genre_count.index, y=genre_count.values, palette='viridis')
plt.title('Phn phi th loi')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sbn.scatterplot(data=df_book, x='ratings_count', y='avg_rating', alpha=0.5)
plt.title('Phn phi Average Rating v Ratings Count')
plt.xlabel('Ratings Count')
plt.ylabel('Average Rating')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sbn.scatterplot(data=df_book, x='reviews_count', y='avg_rating', alpha=0.5)
plt.title('Phn phi Average Rating v Reviews Count')
plt.xlabel('Reviews Count')
plt.ylabel('Average Rating')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
print("Phn phi rating t tp d liu reviews gc:")
original_rating_counts = df_review['rating'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
sbn.barplot(x=original_rating_counts.index, y=original_rating_counts.values, palette='viridis')
plt.title('Phn phi Rating (1-5 sao)')
plt.xlabel('Rating')
plt.ylabel('S lng Review')
plt.xticks(rotation=0)
plt.show()
```

```python
df_merged = pd.merge(df_review, df_book, on='work_id', how='left')
print("Thng tin ca DataFrame sau khi hp nht:")
df_merged.info()
```

```python
reviews_per_author = df_merged.groupby('author').size().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sbn.barplot(x=reviews_per_author.head(10).index, y=reviews_per_author.head(10).values, palette='viridis')
plt.title('Top 10 tc gi c nhiu lt Review nht')
plt.xlabel('Tc gi')
plt.ylabel('S lt Review')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
reviews_per_book = df_merged.groupby('original_title').size().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
sbn.barplot(y=reviews_per_book.head(10).index, x=reviews_per_book.head(10).values, palette='viridis')
plt.title('Top 10 sch c nhiu lt Review nht')
plt.xlabel('S lt Review')
plt.ylabel('Tn sch')

plt.tight_layout()
plt.show()
```
